# Rhonda Hill Geometry

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt


from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)


%config InlineBackend.figure_format = 'retina'
```

```python
# Read 2017 RH dataset
RH_2017 = pd.read_csv('slides2017_RH.csv')
RH_2017.tail()
```

```python
RH_2019 = pd.read_csv('rhondahill_slides.csv')
RH_2019.tail()
```

## slope 

```python
sns.set(font_scale=3)  

slope2019 = np.log10(RH_2019['MEAN'])
slope2017 = np.log10(RH_2017['slope_mean'])

fig = plt.figure()
ax = fig.add_subplot(111)
ax1 = ax.twinx()
ax1.grid(None)
ax1.axis('off')

ax.set_xlim(1,2)
ax.set_ylim(0,1)
#ax1.set_ylim(0, 10)
#ax1.yaxis.set_major_locator(MultipleLocator(0.8))

weightsSlope2019 = np.ones_like(slope2019) / len(slope2019)
weightsSlope2017 = np.ones_like(slope2017) / len(slope2017)

slope_bins = [1,1.1,1.2,1.3,1.4,1.5,1.6,1.7,1.8,1.9,2]
#slope_bins = [15,20,25,30,35,40,45,50]

ax.hist([slope2019,slope2017], bins=slope_bins, weights=[weightsSlope2019,weightsSlope2017], alpha=0.7, label=['2019 Landslides','2017 Landslides'],edgecolor='white', linewidth=1)
ax1 = sns.distplot(slope2019,hist=False,bins=slope_bins, label='2019 Landslides')
ax1 = sns.distplot(slope2017,hist=False,bins=slope_bins, label='2017 Landslides')


ax.set_xlabel('log(Slope (degrees))')
ax.set_ylabel('Fraction of slides')



fig = plt.gcf()
fig.set_size_inches(14.75, 8.75)
#save figure
fig.savefig('figures/rh_slopeplot.pdf', bbox_inches='tight')
```

## failure plane depth

```python
# read in 2019 data
surveydata19 = pd.read_csv('surveydata.csv')
surveydata19.head(3)
```

```python
# read in 2017 data
surveydata17 = pd.read_csv('surveyed_database_2017.csv')
surveydata17 = surveydata17.dropna(subset=['region'])
surveydata17.tail()
```

```python
d2019 = surveydata19['average depth (m)']
d2017 = surveydata17['depth_mean_m']

weightsD2019 = np.ones_like(d2019) / len(d2019)
weightsD2017 = np.ones_like(d2017) / len(d2017)

depth_bins = [0.1, 0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7]

fig = plt.figure()
ax = fig.add_subplot(111)
ax1 = ax.twinx()
ax1.axis('off')

ax.hist([d2019,d2017], bins=depth_bins, weights=[weightsD2019,weightsD2017], alpha=0.75, label=['2019 Landslides','2017 Landslides'],edgecolor='white', linewidth=1)
ax1 = sns.distplot(d2019,bins=depth_bins,hist=False,label='2019 Landslides')
ax1 = sns.distplot(d2017,bins=depth_bins,hist=False,label='2017 Landslides')

plt.legend()
ax.set_xlim(0, 1)

ax.set_ylabel('Fraction of slides')
ax.set_xlabel('Average failure plane depth (m)')

fig = plt.gcf()
fig.set_size_inches(14.75, 8.75)
fig.savefig('figures/rh_depth.pdf', bbox_inches='tight')
```
