# overall rainfall

```python
import numpy as np
import pandas as pd
import matplotlib as mp
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%config InlineBackend.figure_format = 'retina'
mpl.rcParams['font.sans-serif'] = "Arial"
mpl.rcParams['font.size'] = 24
#mpl.rc('xtick', labelsize=20) 
#mpl.rc('ytick', labelsize=20) 

import matplotlib.dates as mdates


from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
rain = pd.read_csv('date_rain_prism.csv')
rain['Date'] =  pd.to_datetime(rain['Date'])
rain.head()
```

```python
rain_dateindex = rain.set_index('Date')
```

```python
rain_yearly = rain_dateindex.resample('Y').sum()
rain_yearly.head()
```

```python
year = mdates.YearLocator()   # every year
year_fmt = mdates.DateFormatter('%Y')

fig = plt.figure()
ax = fig.add_subplot(111)
ax1 = ax.twinx()

ax.xaxis.set_major_locator(year)
ax.xaxis.set_major_formatter(year_fmt)

ax.plot(rain.loc['1981':'2019','ppt (mm)'], color='orange', label='PRISM daily rainfall')
ax1.scatter(rain_yearly['1981-12-31':'2019-12-31'], rain_yearly['ppt (mm)'], color='green', label='PRISM yearly rainfall')

hours = mdates.HourLocator()   # every year

ax.set_xlabel('Timestamp (YEAR)')
ax.set_ylabel('Daily Rainfall (mm)')
ax1.set_ylabel('Yearly Rainfall (mm)')

ax.set_xlim('1981','2019-09')
ax.set_ylim(0,100)
ax1.set_ylim(0,1500)

#major ticks
ax.yaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))

ax1.yaxis.set_major_locator(MultipleLocator(500))
ax1.yaxis.set_major_formatter(FormatStrFormatter('%d'))

#minor ticks
ax.yaxis.set_minor_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(hours)

fig.legend()

fig = plt.gcf()
fig.set_size_inches(14.75, 8.75)
fig.savefig('figures/historical_rainfall.pdf', bbox_inches='tight')
```

```python
year = mdates.YearLocator()   # every year
year_fmt = mdates.DateFormatter('%Y')

fig, ax = plt.subplots()

ax.xaxis.set_major_locator(year)
ax.xaxis.set_major_formatter(year_fmt)

ax.plot(rain_yearly.loc['1981-12-31':'2019-12-31'], color='green', label='PRISM yearly rainfall', style="o")

hours = mdates.HourLocator()   # every year

ax.set_xlabel('Timestamp (YEAR)')
ax.set_ylabel('Yearly Rainfall (mm)')

ax.set_xlim('1981','2019-09')
ax.set_ylim(0,1500)

#major ticks
ax.yaxis.set_major_locator(MultipleLocator(500))
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))

#minor ticks
ax.yaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(hours)

plt.xticks(rotation=270, size=11)
plt.legend()

fig = plt.gcf()
fig.set_size_inches(14.75, 8.75)
fig.savefig('figures/historical_rainfall.pdf', bbox_inches='tight')
```
